```python
#Loading the dataset
from sklearn.datasets import load_boston
```

```python
X , y = load_boston(return_X_y = True)
```

```python
print(X.shape)
print(y.shape)
```

```python
X
```

```python
# Converting the X from array to dataframe
X = pd.DataFrame(X)
X.head()
```

```python
y 
```

```python
# Converting the Y from array to dataframe
y = pd.DataFrame(y)
y.head()
```

```python
# Feature Scaling
sc = StandardScaler()
sc_x = sc.fit_transform(X)
sc_x
```

```python
# Scaled X to a data frame
X = pd.DataFrame(sc_x)
X.head()
```

```python
#Splitting the data to training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test , y_train , y_test = train_test_split(X, y , train_size=0.75 , random_state=101)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Deep Learning Model
# Multilayer Perceptron(MLP)

```python
# API -----> There are 2 types of API's
# Sequential (Connected layer by layer)
# Functional (Connected to any layer randomly)
```

```python
DNN = tf.keras.models.Sequential()
```

```python
# Input data connecting with First Hidden layer (L1)
DNN.add(tf.keras.layers.Dense(units=6 , activation= 'relu'))
```

```python
# Input data connecting with Second Hidden layer (L2)
DNN.add(tf.keras.layers.Dense(units=6 , activation= 'relu'))
```

```python
# Adding a Output layer
DNN.add(tf.keras.layers.Dense(units=1))
```

```python
# Apply Optimization Techinques
# GD , SGD, Mini-batch SGD , SGD with momentum , Adagrade, adadelta, RMSProp,Adam
# Apply measured values using weights

DNN.compile(optimizer="adam", loss='mean_squared_error')
```

```python
# Applying the input dataset along Backpropogation method
DNN.fit(X_train , y_train , batch_size=32 , epochs=500) # Batch size---> Sending the data into 32 batches ,
# epoch---> 1 Forward and 1 Backward propogation is epoch now totaly this will work for 500 times
```

```python
DNN.fit(X_train , y_train , batch_size=32 , epochs=100)
```
